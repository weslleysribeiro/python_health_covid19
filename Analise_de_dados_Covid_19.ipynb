```python
import pandas as pd

confirmed = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
confirmed.head() #CabeÃ§alho
```

```python
confirmed.set_index("Province/State").loc["Anhui"] #loc = localiza pelo indice
```

```python
confirmed["Country/Region"].value_counts()
```

```python
confirmed_by_country = confirmed.groupby("Country/Region").sum()
confirmed_by_country.head()
```

```python
confirmed_by_country.sample(10)
```

```python
confirmed_by_country.loc["Brazil"][2:] #coluna dois em diante (fora index) 
```

```python
import matplotlib.pyplot as plt

#grafico extremamente apertado, estamos sendo longe do adequado ao passar uma mensagem errada. 

plt.figure(figsize=(5,7)) #tamanho do meu grafico
confirmed_by_country.loc["Brazil"][2:].plot()
plt.title("Casos confirmados em 'Brazil'") #title
plt.show() #apenas para mostrar "limpo"
```

```python
#em Data Visualization Ã© fundamental estarmos atento aos DOIS EIXOS.
#estamos comparando do momento inicial atÃ© o momento atual

plt.figure(figsize=(15,7)) #tamanho do meu grafico
confirmed_by_country.loc["Brazil"][2:].plot()
plt.title("Casos confirmados em 'Brazil'") #title
plt.show() #apenas para mostrar "limpo"
```

```python
#estamos comparando a partir de uma coluna 1000 atÃ© o momento atual.

plt.figure(figsize=(15,7)) #tamanho do meu grafico
confirmed_by_country.loc["Brazil"][1000:].plot() #alterando meu eixo x, pegando coluna a partir da n 1000
plt.title("Casos confirmados em 'Brazil'") #title
plt.show() #apenas para mostrar "limpo"
```

```python
#estamos comparando entre dois periodos

plt.figure(figsize=(15,7)) #tamanho do meu grafico
confirmed_by_country.loc["Brazil"][2:].plot()
plt.title("Casos confirmados em 'Brazil'") #title
plt.ylim(10000000, 30000000) #limitando meu eixo y
plt.show() #apenas para mostrar "limpo"
```

```python
#exemplo ruim de manusear o eixo Y

plt.figure(figsize=(15,7)) #tamanho do meu grafico
confirmed_by_country.loc["Brazil"][2:].plot()
plt.title("Casos confirmados em 'Brazil'") #title
plt.ylim(0, 90000000) #limitando meu eixo y
plt.show() #apenas para mostrar "limpo"
```

```python
confirmed_by_country.loc["China"][2:].diff() #diferenÃ§a do dia anterior
```

```python
plt.figure(figsize=(15,7))
new_cases_china = confirmed_by_country.loc["China"][2:].diff().dropna() #tirando os numeros em branco
new_cases_china.plot()
```
